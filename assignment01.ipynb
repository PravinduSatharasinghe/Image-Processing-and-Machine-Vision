```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
img_original = cv.imread(r"E:\Academic\Sem 5\My Sem 5\Image Processing and Machine Vision\assignments\assignment 1\spider.png", cv.IMREAD_COLOR)

#split the image into hue, saturation and value planes
hsv = cv.cvtColor(img_original, cv.COLOR_BGR2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

a = 3600
sigma = 70

def f(x,a):
    b = x + (a/128)*np.exp(-(((x - 128)**2)/(2*(sigma**2))))
    c = min(b, 255)
    return c


#plot the transformation
x = np.linspace(0, 255, 256)
y = np.zeros_like(x)
for i in range(len(x)):
    y[i] = f(x[i],a)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0, 255)


#apply the above transformation to the saturation plane
for i in range(len(s)):
    for j in range(len(s[0])):
        s[i,j] = f(s[i,j],a)

#recombine the three planes
img_corrected_hsv = cv.merge((h, s, v))

#convert the corrected image to the BGR color space
img_corrected_bgr = cv.cvtColor(img_corrected_hsv.astype('uint8'), cv.COLOR_HSV2BGR)

#display the original and corrected images
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv.cvtColor(img_original, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
ax[1].imshow(cv.cvtColor(img_corrected_bgr, cv.COLOR_BGR2RGB))
ax[1].set_title('Corrected Image')
plt.show()

```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
img_original = cv.imread(r"E:\Academic\Sem 5\My Sem 5\Image Processing and Machine Vision\assignments\assignment 1\shells.tif", cv.IMREAD_GRAYSCALE)

# Defining your own function to do histogram equalization
def my_histogram_equalization(img):
    img_equalized = np.zeros_like(img)
    hist = np.zeros(256)
    
    # Calculate histogram
    for i in range(len(img)):
        for j in range(len(img[0])):
            hist[img[i, j]] += 1
    hist = hist / (len(img) * len(img[0]))
    
    # Calculate cumulative distribution function (CDF)
    cdf = hist.cumsum()
    
    # Apply equalization
    for i in range(len(img)):
        for j in range(len(img[0])):
            img_equalized[i, j] = int(cdf[img[i, j]] * 255)
    
    return img_equalized

# Apply the custom histogram equalization function to the original image
img_equalized = my_histogram_equalization(img_original)

# Display the original and equalized images with histograms before and after equalization
fig, ax = plt.subplots(2, 2)
fig.set_figheight(10)
fig.set_figwidth(10)
ax[0, 0].imshow(img_original, cmap='gray')
ax[0, 0].set_title('Original Image')
ax[0, 1].imshow(img_equalized, cmap='gray')
ax[0, 1].set_title('Equalized Image')
ax[1, 0].hist(img_original.ravel(), 256, [0, 256])
ax[1, 0].set_title('Histogram of Original Image')
ax[1, 1].hist(img_equalized.ravel(), 256, [0, 256])
ax[1, 1].set_title('Histogram of Equalized Image')
plt.show()

```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
img_original = cv.imread(r"E:\Academic\Sem 5\My Sem 5\Image Processing and Machine Vision\assignments\assignment 1\jeniffer.jpg", cv.IMREAD_COLOR)

#total number of pixels in the image
MN = len(img_original)*len(img_original[0])

#split the image into hue, saturation and value planes
hsv = cv.cvtColor(img_original, cv.COLOR_BGR2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

#display the three planes
fig, ax = plt.subplots(1, 3)
ax[0].imshow(h, cmap='gray')
ax[0].set_title('Hue')
ax[1].imshow(s, cmap='gray')
ax[1].set_title('Saturation')
ax[2].imshow(v, cmap='gray')
ax[2].set_title('Value')
plt.show()

#apply thresholding to the saturation plane, getting foreground mask
ret, mask = cv.threshold(s, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

#display the mask
fig, ax = plt.subplots()
ax.imshow(mask, cmap='gray')
ax.set_title('Foreground Mask')
plt.show()

#Obtaining foreground only using cv.bitwise, compute histogram
img_foreground = cv.bitwise_and(img_original, img_original, mask=mask)
hist = cv.calcHist([img_foreground], [0], mask, [256], [0, 256])

#obtaining cumulative sum of the histogram
cumulative_sum = np.cumsum(hist)

#defining the formula for histogram equalization
def f(x):
    return 255*cumulative_sum[x]/MN

#histogram-equalizing the foreground
for i in range(len(img_foreground)):
    for j in range(len(img_foreground[0])):
        img_foreground[i,j] = f(img_foreground[i,j])

#extracting background
img_background = cv.bitwise_and(img_original, img_original, mask=cv.bitwise_not(mask))

#adding the foreground and background
img_corrected = cv.add(img_foreground, img_background)

#display the image with histogram-equalized foreground
fig, ax = plt.subplots()
ax.imshow(cv.cvtColor(img_corrected, cv.COLOR_BGR2RGB))
ax.set_title('Corrected Image')
plt.show()


```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
img_original = cv2.imread(r"E:\Academic\Sem 5\My Sem 5\Image Processing and Machine Vision\assignments\assignment 1\einstein.png", cv2.IMREAD_GRAYSCALE)

# Define Sobel kernels for X and Y directions
kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=np.float32)
kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype=np.float32)

# Apply convolution with padding manually
def convolve_with_padding(image, kernel):
    m, n = image.shape
    k_m, k_n = kernel.shape
    
    # Calculate padding sizes
    pad_m = k_m // 2
    pad_n = k_n // 2
    
    # Add padding to the image
    padded_img = np.pad(image, ((pad_m, pad_m), (pad_n, pad_n)), mode='constant')
    
    # Convolution with padding
    result = np.zeros_like(image, dtype=np.float32)
    for i in range(m):
        for j in range(n):
            result[i, j] = np.sum(padded_img[i:i+k_m, j:j+k_n] * kernel)
    
    return result

# Apply convolution with padding
img_sobel_x = convolve_with_padding(img_original, kernel_x)
img_sobel_y = convolve_with_padding(img_original, kernel_y)

# Calculate the gradient magnitude image
img_sobel = np.sqrt(img_sobel_x**2 + img_sobel_y**2)

# Display the original and filtered images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(img_original, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(img_sobel, cmap='gray')
ax[1].set_title('Sobel Filtered Image')
plt.show()




```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#load the original image with its original colors
img_original = cv.imread(r"E:\Academic\Sem 5\My Sem 5\Image Processing and Machine Vision\assignments\assignment 1\flower_q9.jpg", cv.IMREAD_COLOR)

#using grabcut to segment the image and showing the final segmentation mask, foreground image, and the background image
mask = np.zeros(img_original.shape[:2], np.uint8)
bgdModel = np.zeros((1,65), np.float64)
fgdModel = np.zeros((1,65), np.float64)
rect = (0, 0, 300, 300)
cv.grabCut(img_original, mask, rect, bgdModel, fgdModel, 5, cv.GC_INIT_WITH_RECT)
mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
img_segmented = img_original * mask2[:,:,np.newaxis]
img_segmented[mask2 == 0] = [255, 255, 255]
fig, ax = plt.subplots(1, 3)
fig.set_figheight(10)
fig.set_figwidth(10)
ax[0].imshow(mask, cmap='gray')
ax[0].set_title('Segmentation Mask')
ax[1].imshow(cv.cvtColor(img_segmented, cv.COLOR_BGR2RGB))
ax[1].set_title('Foreground Image')
ax[2].imshow(cv.cvtColor(img_original - img_segmented, cv.COLOR_BGR2RGB))
ax[2].set_title('Background Image')
plt.show()

#producing an enhanced image with a substantially blurred background and displaying the original alongside the enhanced image
img_blurred = cv.GaussianBlur(img_original, (51, 51), 0)
img_enhanced = img_original.copy()
img_enhanced[mask2 == 0] = img_blurred[mask2 == 0]
fig, ax = plt.subplots(1, 2)
fig.set_figheight(10)
fig.set_figwidth(10)
ax[0].imshow(cv.cvtColor(img_original, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
ax[1].imshow(cv.cvtColor(img_enhanced, cv.COLOR_BGR2RGB))
ax[1].set_title('Enhanced Image')
plt.show()

```
